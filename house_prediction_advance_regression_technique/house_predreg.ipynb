# ok so outliers has been removed successfully now let's convert data's which has 10 unique values as object dtype to integer or floating point dtype

```python
col_to_conv = [col for col in train_df.columns if train_df[col].nunique() <= 10 and train_df[col].dtype == 'object']
len(col_to_conv)
```

```python
def conv_col(col):
    count = 0 
    k = list(train_df[col].unique())
    Dict = {k[i]: count + i for i in range(len(k))}
    df = train_df[col].replace(Dict)
    return df
def conv_col1(col):
    count = 0 
    k = list(test_df[col].unique())
    Dict = {k[i]: count + i for i in range(len(k))}
    df_test = test_df[col].replace(Dict)
    return df_test

modified_train = pd.DataFrame(map(conv_col,col_to_conv)).T
modified_test= pd.DataFrame(map(conv_col1,col_to_conv)).T
```

## SO far our conversion and deal with missing values is done now let's  figure out the scaling columns and let's scale them

```python
sum(list(modified_train_df[num_cols].isna().sum()))
```

```python
from functools import reduce
def extract_scale_cols(df,cols):
    ex_list=[]
    def Sum(a,b):
        return a+b
    def find_col(cols):
        l=list(df[cols])
        res = reduce(Sum,l[:10])
        return res>=100
    ex_list=list(filter(find_col,cols))
    return ex_list
```

```python
extract_cols = extract_scale_cols(modified_train_df,num_cols)
extract_cols.remove('SalePrice')
```

```python
modified_train_df[extract_cols] = ez.scale(modified_train_df,extract_cols)
modified_test_df[extract_cols] = ez.scale(modified_test_df,extract_cols)
```

let's implement cross-validation and see how good our linearrgression model is

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression,SGDRegressor, ElasticNet, Lasso,Ridge
import numpy as np
linreg = LinearRegression()
```

```python
kf = KFold(n_splits=5, shuffle=True)
cvscores = cross_val_score(linreg, X,y,cv=kf)
print('cross-val scores', cvscores)
print('mean cross-val scores', np.mean(cvscores))
linear_models = [LinearRegression(),SGDRegressor(), ElasticNet(), Lasso(),Ridge()]
```

```python
def cv(model):
    kf = KFold(n_splits=5, shuffle=True)
    cvscores = cross_val_score(model, X,y,cv=kf)
    print('cross-val scores', cvscores)
    print('mean cross-val scores\n', np.mean(cvscores))
```

```python
for model in linear_models:
    cv(model)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(max_depth=30,random_state=42,min_samples_leaf=10)
```

```python
def maxdepth(m):
    model = DecisionTreeRegressor(min_samples_leaf=10,max_depth=m)
    cvscores = cross_val_score(model, X,y,cv=kf)
    return np.mean(cvscores)
```
